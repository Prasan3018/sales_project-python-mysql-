```python
import sqlalchemy as sal

# MySQL connection details
username = "root"
password = "manager"
host = "127.0.0.1"
port = "3306"
database = "code"

# First, connect to MySQL without specifying a database
engine_no_db = sal.create_engine(f'mysql+pymysql://{username}:{password}@{host}:{port}')
conn_no_db = engine_no_db.connect()

# Create the database if it doesn't exist
conn_no_db.execute(sal.text(f"CREATE DATABASE IF NOT EXISTS {database}"))
conn_no_db.close()

# Now, connect to the newly created database
engine = sal.create_engine(f'mysql+pymysql://{username}:{password}@{host}:{port}/{database}')
conn = engine.connect()

print(f"Connected to MySQL database '{database}' successfully!")
```

```python
df.to_sql('df_orders', con=engine, index=False, if_exists='append')
```
